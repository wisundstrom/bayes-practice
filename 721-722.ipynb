```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize':(14,8)})
```

```python
data = pd.read_csv('posterior_gdpInfantMortality.csv')
```

```python
data.dropna(inplace=True)
data.rename(columns={'infant.mortality':'im'}, inplace=True)
```

```python
data.head()
```

```python
linreg=LinearRegression()
X=pd.DataFrame(data.gdp)
y=data.im
linreg.fit(X,y)
```

```python
linreg.score(X,y)
```

```python
X.gdp[0:2]
```

```python
endpoints=np.array([min(X.gdp),max(X.gdp)]).reshape(-1,1)
preds=linreg.predict(endpoints)
```

```python
sns.scatterplot(X.gdp, y);
plt.plot(endpoints,preds,linewidth=2);
```

```python
sns.regplot(X.gdp, y);
```

```python
linreg2=LinearRegression()
X2=pd.DataFrame(np.log(data.gdp))
y2=np.log(data.im)
linreg2.fit(X2,y2)
```

```python
linreg2.score(X2,y2)
```

```python
endpoints2=np.array([min(X2.gdp),max(X2.gdp)]).reshape(-1,1)
preds2=linreg2.predict(endpoints2)
```

```python
sns.scatterplot(x=np.log(data.gdp), y=np.log(data.im));
plt.plot(endpoints2,preds2, linewidth=2)
```

```python
sns.regplot(X2,y2)
```
